next example is a density plot...

```r
install.packages("gridExtra") #just lets you plot subplots inside of a plot. so we can do 4 plots inside 1 plot
library(gridExtra)

common_subplot <- function(data,field) { #you define a function by giving it a name and pushing a value of a function into it. this has 2 parameters coming in, data and field
  ggplot(data,aes(x=get(field), fill=Species))+ #everything between the curly brackets is the logic of the function
    geom_density(alpha=0.4)+ #logic is the ggplot parameter with data and defined aestetic
    labs(x=field, y="Density") #density is a density plot ; alpha makes things more transparent the smaller it is
}

plen<- common_subplot(iris, "Petal.Length")
pwth<- common_subplot(iris, "Petal.Width")
slen <- common_subplot(iris, "Sepal.Length")
swth<- common_subplot(iris, "Sepal.Width")

grid.arrange(plen, pwth, slen, swth, nrow=2) #allows for multiple views
```

```r
theme_set(theme_classic()) #can use atheme within ggplot

g<- ggplot(mpg, aes(displ)) + scale_fill_brewer(palette= "Spectral") #aestetics are the displacement as x axis; scale brewer gives a nice shade

g #outputs a base plot
```

```r
theme_set(theme_classic())

g<- ggplot(mpg, aes(displ)) + scale_fill_brewer(palette= "Spectral")

g + geom_histogram(aes(fill=class), #class adds all the ranges of colors; filling based on class; gives a spectral pallet based on scale
  binwidth = .1,
  col= "black",
  linewidth = .1) +
    labs( #adding labels
      title= "Histogram with Auto Binning", subtitle= "Engine Displacement across vehicle classes"
    )
#higher and higher up I get fewer and fewer so there are a lot more cars with fewer displacement
```

```r
g + geom_histogram(aes(fill=class), #class adds all the ranges of colors
  bins = 5, #5 groupings; grouping based on displacement
  col= "black",
  linewidth = .1) +
    labs(
      title= "Histogram with Fixed Binning", subtitle= "Engine Displacement across vehicle classes"
    )
#visualization plots are all about finding a helpful way to see your data
```
