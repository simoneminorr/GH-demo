```python
import pandas as pd
```

```python
df= pd.read_csv('https://ds1002-resources.s3.amazonaws.com/data/ride_sharing.csv')
#readcsv can look for local files or public ones on the internet
df.head(2)
```

```python
len(df)
#number of rows
```

```python
df.shape
#rows, columns --> dimensions
```

```python
#sort the df
#df.sort_values(by='column_name', ascending=True)
#ascending= True if you want to go from least to greatest
#df. will show you all the commands/functions you can run
#difference between sort and group by: sorting orders the whole data set vs grouping allows you to look at it in subset
df.sort_values(by='station_A_id', ascending=True)
```

```python
df.sort_values(by='station_A_id', ascending=False)
#goes from high to low
#if organizing by names it would go from z to a
```

```python
#view only the records based upon a given value of a given colum
#look at all row
#can filter for string values with .query
df.query('user_gender== "Female"')
```
