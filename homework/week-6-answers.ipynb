# 2. Cleaning/Imputing Data

In class we reviewed methods for removing duplicate rows and handling empty cell values. Download [this data set](https://github.com/nmagee/ds1002/blob/main/data/very-messy-data.csv) and upload into Google Colab.

Create another Pandas DataFrame from that data and perform the following steps.

```python
# Set up pandas and populate the dataframe above using the raw data. Then continue to work with the dataframe in the questions below.
# Fetch the file set, load into colab, and create a new pandas DF.

df2 = pd.read_csv("very-messy-data.csv")
```

```python
# Write a snippet of code that will display the number of duplicate rows exist in the DF.
# First I want to see how many rows are total:
print(len(df2))

# then I want to see how many are dupes
print(df2.duplicated().sum())
```

```python
# Next, remove all duplicate rows using Python.
# There are two ways to do this -- because I want to write out or SAVE the changes.

# Method 1 - assign the same variable name to the modified dataset. This way df2
# will be populated with the version of df2 that has duplicates dropped.
# df2 = df2.drop_duplicates()

# Method 2 - an available parameter of the drop_duplicates method is to tell Python
# to write the changes to the dataframe in place, i.e. save it to the dataframe.
df2.drop_duplicates(inplace=True)
```

```python
len(df2)
```

```python
# For the FOUR columns involving sepal and petal length and width, update all NaN
# values with the column mean such that NO empty cells remain. This method
# is called "imputation" because it fills in cells in a statistically less-impactful
# way than if they were blank.

# Show how you would update those four columns in code

# A little bit of research shows you that you can update multiple columns at the same
# time by specifying them within the df. Here I fill these 4 columns with the column
# mean. Notice that yes you get a FutureWarning.

df2[["sepal_length","sepal_width","petal_length","petal_width"]] = df2[["sepal_length","sepal_width","petal_length","petal_width"]].fillna(df2.mean())
```
